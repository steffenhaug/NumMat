## Problem 1

In this problem we will look at the space of real polynomials of degree $\leq n$:

$$
\mathbb{P}_n(\mathbb{R}) = \left\{f: \mathbb{R}\to\mathbb{R} \quad\biggr\rvert\quad f(x) = \sum_{k=0}^n a_k x^k \quad \textrm{for some }a_1, ..., a_n \in \mathbb{R}\right\}.
$$
Note that each point in the space is a function, namely a polynomial.
### a)
Write down the definition of a real vector space.


<span style="color:blue">
    A set $V$ of vectors equipped with (and closed under) an addition 
    $+:V \times V \rightarrow V$ and a scaling operation 
    $V \times \mathbf R \rightarrow V$ satisfying
    <span>
    <ol>
      <li> $(V, +)$ is an abelian group </li>
      <li> Compatibility of scalar multiplication, $a(b \vec v) = (ab) \vec v$ </li>
        <li>Identity element for scalar multiplication</li>
      <li> Distributive laws for multiplication w. r. t. field and vector addition</li>
    </ol>
    </span>
   is called a real vector space.
</span>

### b)

Show that $\mathbb{P}_n(\mathbb{R})$ is a real vector space.

*Hint: Recall the how we define adding two function, and how we define multiplying with a scalar.*

<span style="color:blue">
    We simply need to check the axioms:
    <span>
    <ol>
      <li> $(\mathbb{P}_n, +)$ is an abelian group
        <ul>
            <li>
                Closure: $f + g = \sum a_k x^k + \sum b_k x^k = \sum (a_k + b_k)x^k \in \mathbb{P}_n$
            </li>
            <li>
                Identity: Clearly the zero function is an additive identity
            </li>
            <li>
                Commutativity: Follows easily from the commutativity in $\mathbf R$. (Since the individual terms in the polynomials commute)
            </li>
            <li>
                Associativity: This follows in the same way.
            </li>
        </ul>
      </li>
      <li> Compatibility of scalar multiplication, $a(b f) = (ab) f$, follows from distributive  and associative laws in $\mathbf R$
          $$a(bf) = a\left(\sum b a_n x^n \right) = ab\left(\sum  a_n x^n \right)$$
      </li>
      <li>Identity element for scalar multiplication is the identity element in $\mathbf R$, namely $1$.
      </li>
      <li> Distributive laws for multiplication w. r. t. field and vector addition
          $$
          c(f + g) = c \sum (a_k + b_k)x^k = \sum (c a_k + c b_k)x^k = cf + cg
          $$
          $$
          (a+b)f = (a+b) \sum c_k x^k = \sum (ac_k+bc_k) x^k = af + bf
          $$
      </li>
    </ol>
    </span>
</span>

### c)

Show that $\mathbb{P}_n(\mathbb{R})$ has dimension $n+1$.

*Hint 1: Find a set of polynomials you think is a basis for $\mathbb{P}_n(\mathbb{R})$. Verify that it is in fact a basis.*

*Hint 2: If $p \in \mathbb{P}_n(\mathbb{R})$ and $p \neq 0$, how many roots can $p$ have at most? Recall the fundamental theorem of algebra.*


<span style="color:blue">
    Candidate: $\{x^k : k \in 0, 1, 2,...n\}$ which as $n+1$ elements.
    Since every element in $\mathbb{P}_n$ is of the form
    $$
    f = \sum c_k x^k
    $$
    this set obviously spans $\mathbb{P}_n$. It is easy to see that the vectors are
    linearly independent (you can not get a quadratic term by scaling and summing linear terms and so on),
    and hence the set is a basis. 
    Since this set has $n+1$ elements, every basis has $n+1$ elements,
    and we say that $\mathbb{P}_n$ has dimension $n+1$.
</span>

### d)
Write down the definition of a norm on a vector space.

<font color ='blue'>
    A norm is a function $||\cdot|| : V \rightarrow \mathbf R _{\geq 0}$ satisfying
    <span>
    <ol>
      <li> $||v|| = 0 \iff v = 0$ </li>
      <li> $||cv|| = |c| ||v||$, where $|c|$ is the absolute value of $c$ in $\mathbf R$ </li>
      <li> $||v + w|| \leq ||v|| + ||w||$ </li>
    </ol>
    </span>
</font>

Let $x_0,...,x_n$ be $n+1$ different real numbers. For a polynomial $q \in \mathbb{P}_n(\mathbb{R})$ we now define
$$
\|q\|_P := \sum_{k=0}^n|q(x_k)|.
$$
Show that $\|\cdot\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R}).$

<font color ='blue'>
    Check the axioms:
    <span>
    <ol>
      <li> $||q|| = 0 \iff q = 0$ because no other polynomial has $n+1$ roots.
      </li>
      <li> $||cq|| = |c| ||q||$
        $$||cq|| = \sum_k \left| c q(x_k) \right| = \sum_k \left| c|| q(x_k) \right| = |c| \sum_k  \left| q(x_k) \right| = |c| ||q|| $$
      </li>
      <li> $||p + q|| = \sum_x |p(x) + q(x)| \leq \sum_x |p(x)| + |q(x)| = ||p|| + ||q||$ </li>
    </ol>
    </span>
</font>
